# Challenge Description

India faces an exceptionally high burden of Type 2 diabetes. India is often referred to as the “diabetes capital of the world,” with over 100 million adults currently living with diabetes, and millions more undiagnosed. While millions are diagnosed, a large proportion of the population remains undiagnosed or diagnosed at advanced stages, often after serious complications such as cardiovascular disease, kidney failure, neuropathy, or vision loss already developed. This is largely due to limited preventive screening, uneven healthcare access, lifestyle changes, and lack of early risk identification. This highlights a critical gap in early detection and preventive care, particularly among urban low-income groups and rural populations.

Type 2 diabetes is a preventable and manageable non-communicable disease (NCD) when detected early. However, current healthcare approaches in India are predominantly reactive rather than preventive, leading to avoidable health complications, increased mortality, and rising economic costs.

The absence of scalable, data-driven early detection systems for diabetes risk in India leads to late diagnosis, preventable complications, and increased healthcare burden. This challenge is not a lack of data, but a lack of intelligent utilization of data for early warning and prevention.


Addressing diabetes through early detection and prevention is critical to achieving long-term health sustainability in India.

This challenge directly aligns with:

SDG 3.4: Reducing premature mortality from non-communicable diseases through prevention and early treatment

SDG 3.8: Strengthening access to quality healthcare services

SDG 3.d: Improving early warning systems for health risks

# Primary Research Question:
Can machine learning models trained on demographic, lifestyle, and clinical data accurately predict the risk of Type 2 diabetes in Indian populations to enable early detection and preventive intervention?

# Hypothesis:
Machine learning–based risk prediction models can identify individuals at high risk of developing Type 2 diabetes more effectively than traditional threshold-based screening methods, enabling earlier intervention.
